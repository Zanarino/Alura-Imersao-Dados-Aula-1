```python
# Importando o pandas e a base de dados usada para aula = Dados sobre imóveis para venda.
import pandas as pd
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
dados.head()
```

```python
#1- Tirar a média das metragens para cada Bairro - Cortando um grupo específico, nesse caso, Bairro e aplicando a média de uma coluna.
dados.groupby('Bairro')['Metragem'].mean()
```

```python
# 2- Duas formas de selecionar os dados por Bairro

# Primeira forma:
dados[dados.Bairro == "Alto de Pinheiros"]
```

```python
# Segunda forma: usando método query
dados.query('Bairro == "Alto da Mooca"')
```

```python
# 3 - Usar outros gráficos da plataforma. Selecionei os imóveis pelo número de vagas disponíveis.

n_imoveis_vagas = dados["Vagas"].value_counts()
n_imoveis_vagas
```

```python
n_imoveis_vagas.head(10).plot.pie()

# Em um gráfico tipo pizza, podemos ver que a maior fatia é de imóveis com 2 vagas. Temos algumas curiosidades com imóveis
# onde há mais vagas disponíveis do que a quantidade de quartos e banheiros.
```

```python
# 4- Pegas outras estatísticas dos dados.

dados.describe()
```

```python
# Podemos pegar também dados isolados com a mediana dos quartos dos imóveis

mediana_quartos = dados['Quartos'].median()
mediana_quartos
```

```python
# 5- Verificar quais os imóveis sem nome de rua

sem_ruas = pd.isna(dados['Rua'])
dados[sem_ruas]
```

```python
# Aqui pegamos os bairros e fazemos a contagem de quantos imóveis estão sem Rua por bairro.

ruas_nan = dados[(dados["Rua"].isnull())]
ruas_nan["Bairro"].value_counts()
```
